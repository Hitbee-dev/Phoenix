```python
import json
import os

with open('/content/drive/MyDrive/korquad_2.0/korquad2.1_train_00.json') as files:
  data = json.load(files)
size = os.path.getsize(f'/content/drive/MyDrive/korquad_2.0/korquad2.1_train_00.json')
with open("/content/drive/MyDrive/korquad_2.0/newfile", "w") as new_file:
  json.dump(data, new_file)

for num in range(1, 39):
  num = str(num).zfill(2)

  with open(f'/content/drive/MyDrive/korquad_2.0/korquad2.1_train_{num}.json') as files:
    data = json.load(files)
  size += os.path.getsize(f'/content/drive/MyDrive/korquad_2.0/korquad2.1_train_{num}.json')
  print(f'File Size({num})', size, 'bytes')
  with open("/content/drive/MyDrive/korquad_2.0/newfile", "a") as new_file:
	  json.dump(data, new_file)

print('File Size_fin', os.path.getsize('/content/drive/MyDrive/korquad_2.0/newfile'), 'bytes')
```
